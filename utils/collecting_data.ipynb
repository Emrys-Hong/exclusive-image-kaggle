```python
import numpy as np
import pandas as pd
import urllib.request
```

```python
training = pd.read_csv('dataset/training.csv')
```

```python
training.head
```

```python
urls = pd.read_csv('dataset/train-images-boxable-with-rotation.csv')
```

```python
len(urls)
```

```python
urls.head
```

```python
urls_id = list(urls.ImageID)
```

```python
len(urls_id)
```

```python
training_len = len(training)
```

```python
for i,o in enumerate(training.loc[:int(training_len/10), 'ImageID']):
    if i%100 == 0:
        print(i)
    url = urls.at[urls_id.index(o), 'OriginalURL']
    urllib.request.urlretrieve(url,'dataset/train/'+o+'.jpg')
```
